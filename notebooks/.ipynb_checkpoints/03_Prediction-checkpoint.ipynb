```python
import numpy

from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential, load_model
from keras.layers import Dense, LSTM

from keract import get_activations, display_activations
import random
```

```python
# PARAMS
#model_file = "/models/model-new.h5" # New Custom Training
model_file = "/models/model-proposed.h5" # 50 neurons + 10 batch + softsign
```

```python
# fix random seed for reproducibility
numpy.random.seed(1234)

clean_file   = "/data/alexa-32k.txt"
malware_file = "/data/dga-32k.txt"
```

```python
# Load clean and malware datasets
with open(clean_file) as f:
    clean_domains = f.read().splitlines()
with open(malware_file) as f:
    malware_domains = f.read().splitlines()
```

```python
# Create Dictionary
charset = list("abcdefghijklmnopqrstuvwxyz0123456789.-")
dictionary = dict(zip(charset, range(len(charset))))
reverse_dictionary = dict(zip(range(len(charset)), charset))
```

```python
# Translate Domain Name to Vector
def domain_to_vector(domain, dictionary):
    res = []
    for c in list(domain):
        v = [float(0)] * len(dictionary)
        v[dictionary[c]] = 1.0
        res.append(v)
    return res
```

```python
# Translate Domain List to DataSet format
def domainlist_to_dataset(domainlist, result, dictionary):
    x = [ domain_to_vector(v, dictionary) for v in domainlist ]
    y = [ [result] for y in range(len(x))]
    
    return x, y
```

```python
# Extract Domain Lists and Merge them
x_clean_noarray, y_clean_noarray = domainlist_to_dataset(clean_domains, 0, dictionary)
x_malware_noarray, y_malware_noarray = domainlist_to_dataset(malware_domains, 1, dictionary)
x_noarray = x_clean_noarray + x_malware_noarray
y_noarray = y_clean_noarray + y_malware_noarray
domains = clean_domains + malware_domains
```

```python
# Padding zeros & Convert to Array
novalue = [float(0)] * len(dictionary)
x_noarray_pad = pad_sequences(x_noarray, dtype=float, value=novalue, padding='post')
x = numpy.array(x_noarray_pad, dtype=float)
y = numpy.array(y_noarray, dtype=float)

print(x.shape)
print(y.shape)
```

```python
# Load LSTM Model
model = load_model(model_file)
model.summary()
```

```python
# Show LSTM Neurons State
#example = ["facebook.com", "facebookgoogleamazonwhateversite.com", "endmq83.com", "dasqwij28ndnas812endmq83.com"]
#example = [(5*"x")+".com", (10*"x")+".com", (15*"x")+".com", (20*"x")+".com", (30*"x")+".com", (40*"x")+".com"]
#example = ['doubleclick.net', 'slideshare.net', 'slickdeals.net', 'adplxmd.com', 'secureserver.net', 'themeforest.net', 'trackingclick.net', 'daikynguyenvn.com', 'prjcq.com', 'bookmyshow.com', 'seesaa.net', 'inquirer.net', 'uploaded.net', 'torcache.net', 'youjizz.com', 'fanfiction.net', 'commentcamarche.net']
example = ["dasqwij28ndnas812endmq83.com", "dasqwijaendnasaioendmque.com", "somethingmalicious.com", "somethingwmallicious.com", "som9ething7mal6ici2ous.com", "supermachinelearningmodels.com"]
example.append(67*"x")
exampleX, exampleY = domainlist_to_dataset(example, 0, dictionary)
exampleP = pad_sequences(exampleX, dtype=float, value=novalue, padding='post')
exampleF = numpy.array(exampleP, dtype=float)

score = numpy.asarray(model.predict(exampleF))
aRes = numpy.where(score >= 0.5, 1, 0)

for i in range(len(example)):
    if aRes[i] == 1:
        res = 'malware'
    else:
        res = 'clean'
    print(example[i], [res]) 
```

```python
for x in exampleF:
    a = get_activations(model, numpy.array([x], dtype=float))
    del a['lstm_1_input']
    del a['dense_1']
    display_activations(a, cmap=None)
```

```python
# Examples of False Negatives
false_negatives = []
for i in range(len(domains)):
    if y[i] != 1:
        continue
        
    example = [domains[i], (63*"x")+".com"]
    exampleX, exampleY = domainlist_to_dataset(example, 0, dictionary)
    exampleP = pad_sequences(exampleX, dtype=float, value=novalue, padding='post')
    exampleF = numpy.array(exampleP, dtype=float)
    
    score = numpy.asarray(model.predict(exampleF))
    aRes = numpy.where(score >= 0.5, 1, 0) 
    res = aRes[0]
        
    if res == 1:
        continue
      
    false_negatives.append(domains[i])
print(len(false_negatives))
```

```python
# Examples of False Positives
false_positives = []
false_positives_top = []
for i in range(len(domains)):
    if y[i] != 0:
        continue
        
    example = [domains[i], (63*"x")+".com"]
    exampleX, exampleY = domainlist_to_dataset(example, 0, dictionary)
    exampleP = pad_sequences(exampleX, dtype=float, value=novalue, padding='post')
    exampleF = numpy.array(exampleP, dtype=float)
    
    score = numpy.asarray(model.predict(exampleF))
    aRes = numpy.where(score >= 0.5, 1, 0) 
    res = aRes[0]
            
    if res == 0:
        continue
      
    false_positives.append(domains[i])
    if i < 1000:
        false_positives_top.append(domains[i])
print(len(false_positives))
```

```python
def print_random(l, n):
    x = list(l)
    random.shuffle(x)
    print(x[0:n])
```

```python
print("### FALSE NEGATIVES")
print_random(false_negatives, 100)
```

```python
print("### FALSE POSITIVES (TOP)")
print(false_positives[0:100])
```

```python
print("### FALSE POSITIVES (RANDOM)")
print_random(false_positives, 100)
```
